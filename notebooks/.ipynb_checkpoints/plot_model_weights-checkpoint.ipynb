```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import sys

# # Add parent directory to sys.path
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
parent_dir = parent_dir + '\\'
sys.path.append(parent_dir)

from config.config import config
```

```python
# Load saved pickle file
with open(parent_dir + 'results\\model_results\\' + 'E40_modelresults.pkl', 'rb') as f:
    data = pickle.load(f)

# Extract required data from the loaded file
coefficients_dict = data['coefficients_dict']
# spike_data = data['spike_data']
r2 = data['r2']
p_values = data['p_values']
lags = config.lags
```

```python
# Define time points
time_points = np.arange(-lags, lags + 1) * config.time_bin  # 0.02s intervals

# Identify valid neurons
valid_neurons = [index for index, value in enumerate(p_values) if value > 0]
select_neurons = valid_neurons
num_valid_neurons = len(select_neurons)
```

```python
# Set up subplots
fig, axes = plt.subplots(
    nrows=np.ceil(num_valid_neurons / 5).astype(int),
    ncols=5,
    figsize=(17, 3 * np.ceil(num_valid_neurons / 5).astype(int))
)
axes = axes.flatten()

# Plot for each valid neuron
for idx, neuron in enumerate(select_neurons):
    ax = axes[idx]
    coefficients_reshaped = coefficients_dict[neuron]
    ax.plot(time_points, coefficients_reshaped[0], label='Run Data', color='red')
    ax.plot(time_points, coefficients_reshaped[1], label='Setpoint Data', color='blue')
    ax.plot(time_points, coefficients_reshaped[2], label='Amplitude Data', color='green')
    ax.axhline(y=0, color='k', linestyle='--')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Coefficient Value / mean FR')
    ax.set_title(f'N {neuron + 1} (R^2: {r2[neuron]:.2f})')
    if idx == 0:  # Add legend to the first subplot
        ax.legend()

# Hide any unused subplots
for idx in range(num_valid_neurons, len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig(parent_dir + 'results\\figures\\' + 'E40_model_weights.png', format='png')
plt.show()
```
